```python
def impute_age(cols):
    Age = cols[1]
    Pclass = cols[0]
    
    if pd.isnull(Age):

        if Pclass == 1:
            return df.groupby('Pclass').mean().loc[1,'Age']

        elif Pclass == 2:
            return df.groupby('Pclass').mean().loc[2,'Age']

        else:
            return df.groupby('Pclass').mean().loc[3,'Age']

    else:
        return Age
```

```python
df['Age']= df[['Pclass','Age']].apply(impute_age,axis=1)
```
